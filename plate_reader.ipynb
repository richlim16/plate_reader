```python
import cv2
from paddleocr import PaddleOCR
from IPython.display import Image
from time import time
import matplotlib.pyplot as plt
```

```python
car_detector = cv2.CascadeClassifier("model/cars.xml")
plate_detector = cv2.CascadeClassifier("model/haarcascade_russian_plate_number.xml")
image = cv2.imread("cars/sample_img_2001.jpg")
```

```python
car = car_detector.detectMultiScale(image, 1.05, 8)
print(car)
for (x, y, w, h) in car:
    cv2.rectangle(image, (x-10, y-10), (x+w, y+h), (0, 255, 255), 2)

plt.imshow(image)
```

```python
plates = plate_detector.detectMultiScale(image, 1.05, 4)

for(a, b, c, d) in plates:
    cv2.putText(image, "Plate", (a,b-5), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, (255, 0, 255), 2)
    cv2.rectangle(image, (a, b), (a+c, b+d), (0, 255, 0), 2)
print(plates)
plt.imshow(image)
```

```python
reader = PaddleOCR(lang='en',show_log=False)
```

```python
output = reader.ocr(image)
output
```
